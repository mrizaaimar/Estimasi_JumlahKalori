# Deskripsi Library yang digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Memanggil Dataset

```python
df = pd.read_csv('starbacs_food.csv')
```

# Visualisasi Data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

# Jumlah Kalori Berdasarkan Kategori

```python
models = df.groupby('Category').count()[['Protein(g)']].sort_values(by='Protein(g)',ascending=True).reset_index()
models = models.rename(columns={'Protein(g)':'KaloriOfKategori'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['Category'], y=models['KaloriOfKategori'], color='royalblue')
plt.xticks(rotation=60)
```

# Distribusi Jumlah Kalori

```python
plt.figure(figsize=(15,5))
sns.distplot(df['Calories'])
```

# Seleksi Fitur

```python
features = ['Protein(g)', 'Total_Fat(g)', 'Sodium(mg)','Total_Carbohydrate(g)', 'Sugars(g)']
x = df[features]
y = df['Calories']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python
#Protein : 12.0 , Total Fat : 5.0 ,Sodium : 530.0 , Total Carbohydrate : 50.0 , Sugars : 5.0
input_data = np.array([[12.0,5.0,530.0,50.0,5.0]])

prediction = lr.predict(input_data)
print('Estimasi jumlah kalori Makanan yang ada di Menu Starbucks yaitu :', prediction, 'kCal')
```
